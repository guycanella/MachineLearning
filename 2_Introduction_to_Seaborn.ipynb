# Introdução ao Seaborn

* Biblioteca de visualização estátística de dados
* Projetada para funcionar em harmonia com os DataFrames

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

# Plots de distribuição
* displot
* jointplot
* pairplot
* rugplot
* kdeplot

```python
tipos = sns.load_dataset('tips')
```

```python
tipos.info()
```

```python
tipos.head()
```

# Distplot
* Plota um histograma da coluna desejada

```python
sns.displot(tipos['total_bill'],kde=True)
```

# Jointplot: distribuição de dados conjuntos para 2 colunas diferentes

```python
sns.jointplot(x='total_bill', y='tip', data=tipos)
#quanto maior a conta, maior a gorjeta
```

```python
sns.jointplot(x='total_bill', y='tip', data=tipos, kind='reg')
```

# Pairplot: pega todas as variáveis numéricas do DataFrame e mostra relação entre elas

```python
sns.pairplot(tipos)
```

```python
# hue: adiciona informações sobre variáveis não numéricas
sns.pairplot(tipos, hue='sex')
```

# Rugplot: distribuição marginal utilizando traços

```python
sns.rugplot(tipos['total_bill'])
```

# Kdeplot:
* A histogram aims to approximate the underlying probability density function that generated the data by binning and counting observations. Kernel density estimation (KDE) presents a different solution to the same problem. Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate:

```python
sns.kdeplot(tipos['total_bill'])
sns.rugplot(tipos['total_bill'])
```

# Plots categóricos
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

# Barplot: representa uma estimativa da tendência central de uma variável numérica através da altura de cada retângulo e fornece uma indicação da incerteza ao redor dessa estimativa
 

```python
sns.barplot(x='sex', y='total_bill', data=tipos)
```

# Countplot: uma simplificação do barplot, onde a função 'estimator' é um contador

```python
sns.countplot(x='sex',data=tipos)
```

# Boxplot: mostra a distribuição de um conjunto de dados quantitativo de uma maneira que facilita a comparação entre as variáveis ou entre os níveis de variáveis categóricas

```python
sns.boxplot(x='day', y='total_bill', data=tipos)
```

```python
sns.boxplot(x='day', y='total_bill', data=tipos, hue='sex')
```

# Violinplot: semelhante ao boxplot mas ao invés de apresentar os componentes do gráfico como pontos de dados reais, o violinplot apresenta um KDE da distribuição subjacente

```python
sns.violinplot(x='day', y='total_bill', data=tipos)
```

```python
sns.violinplot(x='day', y='total_bill', data=tipos, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tipos, hue='sex', split=True)
```

# Stripplot: apresenta um scatterplot onde uma das variáveis é categórica

```python
sns.stripplot(x='day', y='total_bill', data=tipos)
```

```python
sns.stripplot(x='day', y='total_bill', data=tipos, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tipos, hue='sex', dodge=True)
```

# Swarmplot: apresenta um gráfico de dispersão categórico com pontos não sobrepostos

```python
# Lida muito mal com uma grande quantidade de dados

sns.swarmplot(x='day', y='total_bill', data=tipos)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tipos, hue='sex', dodge=True)
```

# Plots matriciais

```python
fly = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

```python
fly.head()
```

```python
tips.head()
```

```python
# Primeiro cria-se uma matriz de correlação entre as variáveis numéricas de um DataFrame
```

```python
crr = tips.corr()
crr
```

```python
sns.heatmap(crr, cmap="magma_r")
```

```python
sns.heatmap(crr, cmap="magma_r", annot=True)
```

```python
# Queremos que:
#- linhas: month
#- colunas: year
#- values: passengers

fly.head()
```

```python
pf = fly.pivot_table(values='passengers', index='month', columns='year')
pf
```

```python
sns.heatmap(pf, cmap='magma_r')
```

```python
sns.clustermap(pf, cmap='magma_r')
```

# PairGrids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.info()
```

```python
iris['species'].value_counts()
```

```python
pg = sns.PairGrid(iris)
pg
```

```python
# map aplica a determinada função em todo o grid ou em uma parte dele
pg.map(plt.scatter)
```

```python
pg = sns.PairGrid(iris)
pg.map_diag(plt.hist)
```

```python
pg = sns.PairGrid(iris)
pg.map_upper(plt.scatter)
pg.map_diag(plt.hist)
pg.map_lower(sns.kdeplot)
```
