# Métodos de K Means Clustering

Métodos de Machine Learning baseados em aprendizado **não supervisionado** que tentará agrupar seus dados em grupos baseados em características similares.

São usados para:
* Agrupamento automático de documentos
* Agrupamento de clientes
* Segmentação de mercado
* Geoestatística

*O objetivos desses métodos são dividir os dados em K grupos distintos baseado nos parâmetros*

**Obs:** esses métodos são iterativos.

<img src="KmeansClustering.png" >

O algoritmo:
* Escolhe um número 'K' de grupos (clusters)
* Defini, aleatoriamente, uma classe para os pontos
* Até os clusters pararem de mudar, faça:
    * Para cada cluster, obtenha o centróide do mesmo calculando a média dos vetores dos pontos do cluster
    * Defina cada ponto ao cluster na qual o centróide é o mais próximo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data[0][:,0], data[0][:,1], c=data[1], s=100)
```

```python
from sklearn.cluster import KMeans
```

```python
# n_clusters=4 porque sabemos que temos 4 conjuntos de dados diferentes

kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
# Posições dos centróides dos grupos

kmeans.cluster_centers_
```

```python
# Labels que o modelo verificou

kmeans.labels_
```

```python
fig, ax = plt.subplots(1,2,sharey=True, figsize=(12,8))

ax[0].set_title('KMeans')
ax[0].scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')
ax[1].set_title('Original')
ax[1].scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
